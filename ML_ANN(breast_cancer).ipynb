```python
#importing necessary libraries
import keras
import pandas as pd
import numpy as np
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wdbc.data"
data = pd.read_csv(url, header=None)
```

```python
data
```

```python
#splitting dataset
data.drop(0, axis=1, inplace=True)
X = data.iloc[:, 1:]
y = data.iloc[:, 0]
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
y = encoder.fit_transform(y)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation


model = Sequential()
model.add(Dense(10, activation='sigmoid', input_dim=X_train.shape[1]))
model.add(Dense(10, activation='softmax'))
model.add(Dense(1, activation='sigmoid'))  # Output layer for binary classification
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy',
              optimizer="sgd",
              metrics = ['accuracy'])
#loss is binary crossentropy since we only have two labels....
```

```python
model.fit(X_train, y_train, batch_size=100, epochs=10)
```

```python
model.fit(X_train, y_train, batch_size=100, epochs=25)
```

```python
model.fit(X_train, y_train, batch_size=100, epochs=250)#constant accuracy may be due to vanishing gradient
```

```python
test_loss, test_acc = model.evaluate(X_test, y_test)
```

```python
print(f'Test Loss: {test_loss}, Test Accuracy: {test_acc}')
```

```python
predictions = model.predict(X_test)
np.argmax(predictions[9])
print(predictions[9])
np.sum(predictions[9])
import numpy as np

predicted_class = np.argmax(predictions[9])
actual_class = y_test[9]

if predicted_class == actual_class:
    print("Prediction is correct!")
else:
    print("Prediction is incorrect.")
```
